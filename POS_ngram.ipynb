```python
import spacy
import nltk
import math
```

```python
pos_txt = []
neg_txt = []
with open("train.txt", encoding="cp863") as txt:
    lines = txt.read().splitlines()
    for line in lines:
        parts = line.split("\t")
        pos_txt.append(parts[0])
        neg_txt.append(parts[1])
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def output(sent):
    pos_bigrams = []
    tag_bigrams = []
    doc = nlp(sent)

    for i in range(len(doc)-1):
        pos_bigrams.append(f"{doc[i].pos_}_{doc[i+1].pos_}")
        tag_bigrams.append(f"{doc[i].tag_}_{doc[i+1].tag_}")
    return pos_bigrams, tag_bigrams
        #print(token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_)
```

```python

train_pos = pos_txt[:round(len(pos_txt) * 0.8)]
val_pos = pos_txt[round(len(pos_txt) * 0.8):]
train_neg = neg_txt[:round(len(neg_txt) * 0.8)]
val_neg = neg_txt[round(len(neg_txt) * 0.8):]
```

```python
vocab = {}
for i in range(len(train_pos)):
    pos_words = nltk.tokenize.word_tokenize(train_pos[i])
    for word in set(pos_words):
        if word not in vocab:
            vocab[word] = 1
        else:
            vocab[word] += 1
```

```python
wrong_spell = {}
for i in range(len(train_pos)):
    neg_words = nltk.tokenize.word_tokenize(train_neg[i])
    pos_words = nltk.tokenize.word_tokenize(train_pos[i])
    for word in set(neg_words):
        if word not in vocab:
            if word not in wrong_spell:
                wrong_spell[word] = 1
            else:
                wrong_spell[word] += 1
```

```python
total_spell = sum(wrong_spell.values())
```

```python
spell_dict = {}
for w in wrong_spell:
    if w not in spell_dict:
        spell_dict[w] = math.log(wrong_spell[w] / total_spell)
```

```python
spell_dict
```

```python
wrong_pos_bigrams = {}
wrong_tag_bigrams = {}
for i in range(len(train_pos)):
    pos_post, tag_post = output(train_pos[i])
    pos_neg, tag_neg = output(train_neg[i])
    diff_pos = set(pos_neg).difference(set(pos_post))
    diff_tag = set(tag_neg).difference(set(tag_post))
    for bi in diff_pos:
        if bi not in wrong_pos_bigrams:
            wrong_pos_bigrams[bi] = pos_neg.count(bi)
        else:
            wrong_pos_bigrams[bi] += pos_neg.count(bi)
    for tag in diff_tag:
        if tag not in wrong_tag_bigrams:
            wrong_tag_bigrams[tag] = tag_neg.count(tag)
        else:
            wrong_tag_bigrams[tag] += tag_neg.count(tag)
    if i % 20000 == 0:
        print(f"have been finished {i} sentences...")
```

```python
def tri_output(sent):
    pos_trigrams = []
    tag_trigrams = []
    doc = nlp(sent)

    for i in range(len(doc)-2):
        pos_trigrams.append(f"{doc[i].pos_}_{doc[i+1].pos_}_{doc[i+2].pos_}")
        tag_trigrams.append(f"{doc[i].tag_}_{doc[i+1].tag_}_{doc[i+2].tag_}")
    return pos_trigrams, tag_trigrams
```
